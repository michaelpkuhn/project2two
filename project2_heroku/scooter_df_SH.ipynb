# Average Weekday/Weekend Rides

```python
# Calculate Average Trips per Hour For weekdays and weekends

hour_rides = scooter.groupby(['Day_End','Time'])['Trip_Count'].sum().reset_index()
hour_rides

# hour_rides.dtypes


# Add data for 3am
new_row = {'Time': '03:00:00 AM', 'Day_End':'Weekday', 'Trip_Count': 0}
hour_rides = hour_rides.append(new_row, ignore_index=True)
new_row= {'Time': '03:00:00 AM', 'Day_End':'Weekend', 'Trip_Count': 0}
hour_rides = hour_rides.append(new_row, ignore_index=True)

# Add data for 4am
new_row = {'Time': '04:00:00 AM', 'Day_End':'Weekday', 'Trip_Count': 0}
hour_rides = hour_rides.append(new_row, ignore_index=True)
new_row= {'Time': '04:00:00 AM', 'Day_End':'Weekend', 'Trip_Count': 0}
hour_rides = hour_rides.append(new_row, ignore_index=True)

# Sort Times
hour_rides = hour_rides.iloc[pd.to_datetime(hour_rides.Time).values.argsort()]
hour_rides
```

```python
# Add column with Weekday and weekend count
hour_rides['Day_Count'] = [76 if ele == 'Weekday' else 32 for ele in hour_rides['Day_End']]

# Calculate average rides per hour on Weekdays and Weekend
hour_rides['Avg_Per_Hour'] = (hour_rides['Trip_Count']/hour_rides['Day_Count']).round()
hour_rides


hour_rides
# hour_rides.dtypes
```

```python
# Convert and save Trip Count by Date as json file
hour_rides.to_json(r'data/avg_hr_rides.json',orient='records')
```

# Trip Distance

```python
# Trip Count per Community
comm_trips = scooter.groupby(['Start_Community_Name'])['Trip_Count'].sum().reset_index()
# comm_trips

# Sort Trip Count in Descending order
comm_trips =comm_trips.sort_values(by='Trip_Count', ascending=False)
comm_trips
```

```python
# Bottom Community list
bottom= ['AVONDALE','BELMONT CRAGIN','LOWER WEST SIDE', 'HERMOSA','AUSTIN', 'EAST GARFIELD PARK','HUMBOLDT PARK','IRVING PARK','NORTH LAWNDALE', 'SOUTH LAWNDALE','PORTAGE PARK', 'WEST GARFIELD PARK', 'MONTCLARE','DUNNING', 'NEAR NORTH SIDE', 'BRIDGEPORT','LINCOLN PARK']
mid = ['LOGAN SQUARE', 'AVONDALE', 'BELMONT CRAGIN', 'LOWER WEST SIDE', 'HERMOSA', 'AUSTIN', 'EAST GARFIELD PARK', 'HUMBOLDT PARK', 'IRVING PARK']
```

```python
# Restructure scooter dataframe
comm_trips = scooter.filter(items=['Start_Community_Name','Distance','Trip_Count'])
comm_trips
```

```python
def f(row):

    if row['Distance'] >5:
        val = ">5 miles"
    elif row['Distance'] >= 4:
        val = "4-5 miles"
    elif row['Distance'] >= 3:
        val = "3-4 miles"
    elif row['Distance'] >= 2:
        val = "2-3 miles"
    elif row['Distance'] >= 1:
        val = "1-2 miles"
    else:
        val =  "<1 mile"

    return val



# def f(row):

#     if row['Distance'] >20:
#         val = ">20 miles"
#     elif row['Distance'] >= 15:
#         val = "20-15 miles"
#     elif row['Distance'] >= 10:
#         val = "10-15 miles"
#     elif row['Distance'] >=5:
#         val = "5-10 miles"
#     else:
#         val =  "<5 miles"
```

```python
comm_trips["Distance_Range"] = comm_trips.apply(f,axis=1)
comm_trips
```

# Top Community Trip Distances

```python
# Drop Bottom Ride Communities 
top_comm = comm_trips[~scooter['Start_Community_Name'].isin(bottom)]
top_comm
```

```python
# Groupby top Community and Distance range with trip count per range

comm_dist = top_comm.groupby(['Start_Community_Name','Distance_Range'])['Trip_Count'].sum().reset_index()
comm_dist
```

```python
# Define a function to map the grand total trip values for top 3 communities 
def set_value(row_number, assigned_value): 
    return assigned_value[row_number] 
  
# Create the dictionary 
event_dictionary ={'LOGAN SQUARE' : 34732, 'WEST TOWN' : 92666, 'NEAR WEST SIDE' : 110849} 
  
# Add a new column named 'Total Rides' 
comm_dist['Total_Rides'] = comm_dist['Start_Community_Name'].apply(set_value, args =(event_dictionary, )) 
  
# Print the DataFrame 
comm_dist 
```

```python
# Divide Distange Range-Trip Count by Total Rides & add "Trip_Pct" column

comm_dist['Trip_Pct'] = ((comm_dist['Trip_Count']/comm_dist['Total_Rides'])*100).round()
comm_dist
```

```python
# Convert and save dataframe as json file
comm_dist.to_json(r'data/comm_dist.json',orient='records')
```
